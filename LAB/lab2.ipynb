```python
# Write a program to find the following between two vectors a = [2,3,7,8,7]
# and b = [4,7,8,2,1].
# a. Euclidean distance
# b. Manhattan distance
# c. Cosine Similarity
```

```python
a = np.array([2,3,7,8,7])
b = np.array([4,7,8,2,1])

np.sqrt(np.sum((a-b)**2))
```

```python
a = np.array([2,3,7,8,7])
b = np.array([4,7,8,2,1])

np.sum(np.abs(a-b))
```

```python
a = np.array([2,3,7,8,7])
b = np.array([4,7,8,2,1])

dot_product = np.dot(a,b)

magnitude_a = np.sqrt(np.sum(a**2))
magnitude_b = np.sqrt(np.sum(b**2))

cosine_similarity = dot_product / (magnitude_a * magnitude_b)
print(cosine_similarity)
```

```python
# Write a program to find a simple matching coefficient and Jaccard similarity between [1,0,1,0,0,0] and [1,0,1,1,0, 1].

a = np.array([1, 0, 1, 0, 0, 0])
b = np.array([1, 0, 1, 1, 0, 1])

# Simple Matching Coefficient (total matches / total items)
matches = np.sum(a == b)
smc = matches / len(a)

# Jaccard Similarity (intersection/union of 1's)
intersection = np.sum((a == 1) & (b == 1))
union = np.sum((a == 1) | (b == 1))
jaccard = intersection / union

print("Simple Matching Coefficient:", smc)
print("Jaccard Similarity:", jaccard)
```

```python
# 3. Write a program to perform feature scaling (min max, z score and decimal scaling) on [35000, 42000, 39000, 45000]

x = np.array([35000, 42000, 39000, 45000], dtype=float)

x_min = np.min(x)
x_max = np.max(x)
x_new = (x-x_min) / (x_max - x_min)

mean = np.mean(x)
std = np.std(x)
z_score = (x - mean) / std

j = len(str(int(np.max(np.abs(x))))) 
decimal = x / (10**j)

print(x)
print(x_new)
print(z_score)
print(decimal)
```

```python
# You are provided with the following arrays representing the monthly sales data (in USD) of two products - Laptops and Smartphones - sold by an electronics store over the first 6 months of the year:
# months = ['January', 'February', 'March', 'April', 'May', 'June']
# laptop_sales = [12000, 15000, 17000, 13000, 16000, 18000]
# smartphone_sales = [10000, 14000, 16000, 12000, 15500, 17500]


# a. Plot the sales trends of both laptops and smartphones over the 6 months.

import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June']
laptop_sales = [12000, 15000, 17000, 13000, 16000, 18000]
smartphone_sales = [10000, 14000, 16000, 12000, 15500, 17500]

plt.plot(months, laptop_sales, marker='^', label='Laptops')
plt.plot(months, smartphone_sales, marker='s', label='Smartphones')
plt.title('Sales Trends Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# b. Create a scatter plot comparing laptop sales vs smartphone sales.

plt.scatter(laptop_sales, smartphone_sales, color='purple')
for i, month in enumerate(months):
    plt.text(laptop_sales[i]+200, smartphone_sales[i]+200, month)
plt.title('Laptop Sales vs Smartphone Sales')
plt.xlabel('Laptop Sales (USD)')
plt.ylabel('Smartphone Sales (USD)')
plt.grid(True)
plt.show()
```

```python
# c. Draw a bar chart showing sales of laptops and smartphones for each month.

x = np.arange(len(months))
width = 0.35

plt.bar(x - width/2, laptop_sales, width, label='Laptops', color='skyblue')
plt.bar(x + width/2, smartphone_sales, width, label='Smartphones', color='orange')
plt.xticks(x, months)
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.legend()
plt.show()
```

```python
# d. Plot a pie chart showing the percentage contribution of each month to total laptop sales.
plt.pie(laptop_sales, labels=months, autopct='%1.1f%%', startangle=140, colors=plt.cm.Blues(np.linspace(0.3, 0.7, len(months))))
plt.title('Percentage Contribution of Each Month to Total Laptop Sales')
plt.show()
```

```python
# e. Plot another pie chart for smartphone sales.
plt.pie(smartphone_sales, labels=months, autopct='%1.1f%%', startangle=140, colors=plt.cm.Oranges(np.linspace(0.3, 0.7, len(months))))
plt.title('Percentage Contribution of Each Month to Total Smartphone Sales')
plt.show()
```
